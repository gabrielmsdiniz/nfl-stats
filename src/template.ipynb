```python
# Importing packages
import pandas as pd
import numpy as np
```

```python
selected_data = ['Player Id', 'Year', 'Team', 'Games Played','Passes Attempted', 'Completion Percentage',
'Passing Yards', 'Passing Yards Per Game',
'Int Rate', 'Sacks', 'Passer Rating']

pass_stats = pd.read_csv('data/Career_Stats_Passing.csv')
pass_stats = pass_stats[selected_data]
pass_stats.head()
```

```python
pass_stats = pass_stats[pass_stats.Year >= 2000]
pass_stats = pass_stats[pass_stats.Year <= 2016]
pass_stats = pass_stats[pass_stats['Games Played'] > 0]
```

```python
# Format to int values

to_int_columns = ['Games Played', 'Passes Attempted', 'Sacks']

for col in to_int_columns:
    pass_stats[col] = pass_stats[col].replace('--', None)
    pass_stats[col] = pass_stats[col].astype(int)

pass_stats.dtypes
```

```python
# Format to float values
to_float_columns = ['Completion Percentage', 'Passing Yards', 'Passing Yards Per Game', 'Int Rate', 'Passer Rating']


for col in to_float_columns:
    pass_stats[col] = pass_stats[col].replace('--', None)
    pass_stats[col] = pass_stats[col].replace(',', '.', inplace=True)
    pass_stats[col] = pass_stats[col].astype(float)

pass_stats.dtypes
```
